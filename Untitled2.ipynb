```python
# Import necessary libraries
from sklearn.linear_model import LogisticRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Use the raw content URL for the CSV file
df = pd.read_csv("https://raw.githubusercontent.com/ibm-ice/Data-Exchange/main/diabetes.csv")
df
```

```python
# make the x(independent variable) and y(dependent variable) differen
x=df[["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]]#indedendent or input variable
y=df['Outcome']#dependent or outcome variable
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression # Changed to LogisticRegression
from sklearn.metrics import classification_report,accuracy_score # Corrected import statement
```

```python
#traing and testing
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Create and train the Logistic Regression model
model = LogisticRegression()  # Initialize the model
model.fit(x_train, y_train)    # Train the model using training data
```

```python
# Make predictions on the test set
y_pred = model.predict(x_test) # Predict using the trained model
```

```python
# Evaluate the model - Now using classification_report and accuracy_score
from sklearn.metrics import classification_report,accuracy_score
# print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
# print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
def core(dataset,threshold):
  col_corr=set()
  corr_matrix=dataset.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range(i):
      if abs(corr_matrix.iloc[i,j])>threshold:
        colname=corr_matrix.columns[i]
        col_corr.add(colname)
  return col_corr
```
